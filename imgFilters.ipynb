**<h2>Question 1</h2>**
Convolution by using conv2d function

```
img = cv2.imread("shoes.png", 0)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
#img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img)
```

```
#Gaussian Blurr
k = np.array(np.ones((11, 11), np.float32))/121

output = cv2.filter2D(img, -1, k)

plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(output)
plt.title('Filtered Image')
```

```
#Edge detection
k = np.array(([-1, -1, -1], [-1, 8, -1], [-1, -1, -1]), np.float32)

output = cv2.filter2D(img, -1, k)

plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(output)
plt.title('Filtered Image')
```

```
#Sharpen
k = np.array(([0, -1, 0], [-1, 5, -1], [0, -1, 0]), np.float32)

output = cv2.filter2D(img, -1, k)

plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(output)
plt.title('Filtered Image')
```

b) Building edge detector by using libraries

```
img = cv2.imread('parrots.jpg')
plt.imshow(img)
```

```
#canny edge detection
#provide the image, 2 main parameters, threshold1 and threshold2 
#we're gonna start off by setting the middle value of what pixel can be (0-225)
edges = cv2.Canny(image=img, threshold1=127, threshold2=127)
plt.imshow(edges)
```

```
edges = cv2.Canny(image=img, threshold1=0, threshold2=255)
plt.imshow(edges)
```

```
#formula to calculate the median pixel value
med_val = np.median(img)
```

64.0

```
# LOWER THRESHOLD TO EITHER 0 OR 70% OF THE MEDIAN VALUE, WHICHEVER IS GREATER
lower = int(max(0, 0.7*med_val))
#UPPER THRESHOLD TO EITHER 130% OF THE MEDIAN VALUE OR THE MAX VALUE 255, WHICHEVER IS SMALLER
upper = int(min(255, 1.3*med_val))
```

```
blurred_img = cv2.blur(img,ksize=(5,5))
```

```
edges = cv2.Canny(image=blurred_img,threshold1=lower,threshold2=upper+50)
plt.imshow(edges)
```

b) edge detector by using libraries

```
from google.colab.patches import cv2_imshow
import cv2

img = cv2.imread('shoes.png', cv2.IMREAD_UNCHANGED)

# Calcution of Sobelx horizontal
sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=5)
    
# Calculation of Sobely vertical
sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=5)
    
# Calculation of Laplacian 
laplacian = cv2.Laplacian(img, cv2.CV_64F)
    
cv2_imshow(sobelx)
cv2_imshow(sobely)
cv2_imshow(laplacian)
```

**<h2>Question 3</h2>**

BINARY THRESHOLDING:

is basically a method of segmenting an image into different parts. It converts an image to 2 values(white or black)

```
img = cv2.imread("thr1.jpg", 0)
plt.imshow(img, cmap='gray')
```

The way threshold works

```
# 2 parameters, actual threshold and max value. any val below a specific threshold get turns to 0
#                                           and any value above that get turns to the max
ret, thresh1 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
```

```
ret
```

```
plt.imshow(thresh1,cmap='gray')
```

THRESHOLD BINARY INVERSE

```
ret, thresh1 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY_INV)
plt.imshow(thresh1,cmap='gray')
```

ADAPTIVE THRESHOLDING

```
th2 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 8)
plt.imshow(th2)
```
